```python
# currentPageNo = 1
Cname1 = []
Ccount1 = []
addpage1 = []
headers1 = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'} # 크롬 webdriver 안쓸거면 이거 해줘야 하는듯ㅅㅂ

url1 = 'https://www.chargekorea.com/charge/list.php?pagenum=27&addcode=11500&searchkeyword=&orderby=count&hashtag=&usenow=&parkingfee=&chargespeed='

r = requests.get(url1, headers=headers1)
# time.sleep(1)
soup = BeautifulSoup(r.text, 'html.parser')
# print(len(soup))

# Cname = []
for aa in range (1,10):
    crawling4 = soup.find_all('div', {'style':'float:left'})[aa].find_all('a')[0].text
    print(crawling4)
    Cname1.append(crawling4)
    
    


for bb in range (0,9):
    crawling5 =soup.find_all('div', {'style':'text-align:center;color:gray;font-size:15px'})[bb].text
    print(crawling5)
    Ccount1.append(crawling5)




for cc in range(0,9):
    commurl = 'https://www.chargekorea.com/charge/'
    crawling6 = soup.find_all('td', {'style':'width:70%'})[cc].select('td>div>a')[0]['href']
    add_url1=commurl+crawling6
    print(add_url1)
    addpage1.append(add_url1)
    addpage1

f = pd.DataFrame({'충전소이름':Cname1})
g = pd.DataFrame({'전체충전기':Ccount1})
h = pd.DataFrame({'충전소주소url':addpage1})

i = pd.concat([f, g, h], axis=1)
i
```
